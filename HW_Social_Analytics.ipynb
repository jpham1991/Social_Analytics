```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import tweepy
import time
import seaborn as sns
import os

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
file_name = "/Users/JENNIFERPHAM/Desktop/api_keys.json"
data = json.load(open(file_name))

consumer_key = data['twitter_consumer_key']
consumer_secret = data['twitter_consumer_secret']
access_token = data['twitter_access_token']
access_token_secret = data['twitter_access_token_secret']
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
bbc_user = "@bbc"

bbc_counter = 1
oldest_tweet = ""

# Variables for holding sentiments
bbc_sentiments = []
bbc_compound_list = []
bbc_positive_list = []
bbc_negative_list = []
bbc_neutral_list = []

for x in range(5):
    
    # Get all tweets from home feed
    bbc_public_tweets = api.search(bbc_user, count=100, result_type="recent", max_id=oldest_tweet)
    # Loop through all tweets 
    for tweet in bbc_public_tweets["statuses"]:

        bbc_compound = analyzer.polarity_scores(tweet["text"])["compound"]
        bbc_pos = analyzer.polarity_scores(tweet["text"])["pos"]
        bbc_neu = analyzer.polarity_scores(tweet["text"])["neu"]
        bbc_neg = analyzer.polarity_scores(tweet["text"])["neg"]
        bbc_compound_list.append(bbc_compound)
        bbc_positive_list.append(bbc_pos)
        bbc_negative_list.append(bbc_neg)
        bbc_neutral_list.append(bbc_neu)
        # Add sentiments for each tweet into an array
        bbc_sentiments.append({"Date": tweet["created_at"], 
                           "Compound": bbc_compound,
                           "Positive": bbc_pos,
                           "Negative": bbc_neg,
                           "Neutral": bbc_neu,
                           "Tweets Ago": bbc_counter})
        
        # Add to counter 
        bbc_counter = bbc_counter + 1
bbc_sentiments_df = pd.DataFrame.from_dict(bbc_sentiments)
bbc_sentiments_df.head()
```

```python
bbc_sentiments_df.to_csv("bbc_sentiments.csv", encoding="utf-8", index=False)
```
